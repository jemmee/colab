```python
from jax.sharding import Mesh, NamedSharding, PartitionSpec as P

# Setup an 8-core mesh
devices = jax.devices()
mesh = Mesh(devices, ('x',))

# Define how to split the data
sharding = NamedSharding(mesh, P('x'))

# Create a giant array sharded across all cores
big_x = jax.device_put(jax.random.normal(jax.random.PRNGKey(0), (8192, 8192)), sharding)

print(f"Array sharding: {big_x.sharding}")
```
